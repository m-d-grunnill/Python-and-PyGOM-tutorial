# 1. Numpy

At the core of numpy is the ndarray object. Although having similarities to lists (and tupples) in terms of indexing, Ndarrays have some key differences:
* They are of a fixed sized from creation (similar to a tupple). Changing the deminsions of a ndarray results in the creation of a new ndarray.
* They facilitate mathmatical operations at a much faster speed and with less code.

This final point has led to numpy array's being the basis of many python packages in mathematical and scientific computing. As such python based work in this area requires often requires knowledge of numpy arrays.

# 1.1 Creating Arrays.

Lets import numpy and create some ndarrays.

```python
import numpy as np # numpy is often shotened to np in coding:
```

```python
array_1d = np.array([3,-1, 11, 3.14])
array_2d = np.array([[ 3, -1 ],
                     [11, 3.14]])
array_3d = np.array([[[ 3, -1 ,2], [2,3.14,3]], [[11, 3.14, 8], [10,11,12]]])

display(array_1d,array_2d,array_3d)
```

Arrays of sequences can be generated using `arange` in similar manner to `range`. **Note** this has an advantage over `range` in that it can handle floats.

```python
display(np.arange(2,5), np.arange(2,3,0.25))
```

There are several methods for creating arrays of repeated values:

```python
ones = np.ones((3, 4))
zeros = np.zeros([1,2])
trues = np.full((2,3),True)
display(ones,zeros,trues)
```

# 1.1 Array shape and indexing.

An arrays deminsions can be determined using `.shape`:

```python
display(array_1d.shape,array_2d.shape,array_3d.shape)
```

When indexing deminsions are indexed in the same order they are created. This means rows first then columns:

```python
display(array_1d[3], array_2d[1,0], array_3d[1, 1,0])
```

As well as slices `:` multiple single elements can be indexed:

```python
display(array_1d[:3], array_2d[[1,0],[0,1]])
```

## 1.3 Element wise operations.

Elementwise operations are accessed via the same methods as carried out on floats and ints:

```python
array_2d_b = np.array([[ 8, -5 ],
                       [10, 6]])
addition_array = array_2d + array_2d_b
sub_array = addition_array - array_2d_b
mul_array = array_2d * array_2d_b
div_array = mul_array/ array_2d_b
display(addition_array, sub_array, mul_array, div_array)
```

# 5. Sympy

Sympy stands for **sym**bolic algebra in **py**thon. Below is an example of Sympy being used to derive $R_0$ (Basic Reproductive Number), following section 2.1 of:

Diekmann, O., Heesterbeek, J. A. P., & Roberts, M. G. (2010). The construction of next-generation matrices for compartmental epidemic models. Journal of the Royal Society Interface, 7(47), 873–885. https://doi.org/10.1098/rsif.2009.0386

```python
#Set up sympy to display outputs as mathmatical formula as opposed to latex.
from sympy.interactive import printing
printing.init_printing(use_latex=True)
```

```python
# First we need to import sympy
import sympy 
# We also need to import copy for copying items
import copy
```

## 5.1 Setup Model

```python
# Setup Symbols for Sympy
N, S, E_1, E_2, I, R, mu, beta, p, nu_1, nu_2, gamma = sympy.symbols('N S E_1 E_2 I R mu beta p nu_1 nu_2 gamma')
```

```python
# The system can be described as
ODE_S = mu*N-beta*S*I/N-mu*S
ODE_E_1 = p*beta*S*I/N-(mu+nu_1)*E_1
ODE_E_2 = (1-p)*beta*S*I/N-(mu+nu_2)*E_2
ODE_I = nu_1*E_1+nu_2*E_2-(mu+gamma)*I
ODE_R = gamma*I-mu*R

ODEs = sympy.Matrix([ODE_S,ODE_E_1,ODE_E_2,ODE_I,ODE_R])
ODEs
```

## 5.2 Get jacobian of infected stages

```python
# We are only interested in the, latent and infectious compartments
ODEs_infecteds = sympy.Matrix([ODE_E_1,ODE_E_2,ODE_I])
ODEs_infecteds
```

## 5.3 Obtainning matrix  of transitions between or out of infectious stages ($\Sigma$)

```python
# The jacobian of the infecteds
jacobian_of_infecteds = ODEs_infecteds.jacobian(X=[E_1, E_2, I])
jacobian_of_infecteds
```

```python
# e.g. removing people becoming infected from the jacobian above.
Sigma = copy.deepcopy(jacobian_of_infecteds)
Sigma[0,2] = 0 # Sigma[0,2] = Sigma[0,2]-S*beta*p/N would work to. .subs method can also be used see below.
Sigma[1,2] = Sigma[1,2]-S*beta*(1-p)/N # Sigma[1,2] = 0 would work to. .subs method can also be used see below.
Sigma
```

## 5.4 Obtainning matrix  of transitions into of infectious stages (T)

```python
# E.g. removing people transitioning from the jacobian above.
# Suggest not use T to name a variable could be confused with transpose of a matrix.
T_inf_births = jacobian_of_infecteds.subs({N:1, S:1, mu:0, nu_1:0, nu_2:0, gamma:0})
T_inf_births
```

## 5.5 Obtainning Next Geneation Matrix (K<sub>L</sub>).

```python
Sigma_inv = Sigma**-1 # note for powers in python it is ** not ^.
neg_Sigma_inv = -Sigma_inv
```

```python
K_L = T_inf_births*neg_Sigma_inv
K_L
```

## 5.6 Finally the Basic Reproductive Number (R<sub>0</sub>)

```python
eigen_values = K_L.eigenvals()
eigen_values
```

```python
none_zero_eigen_values = [item for item in eigen_values.keys() if item !=0]
```

```python
sympy_R0 = none_zero_eigen_values[0]
sympy_R0
```

Lets just sense check that against the Diekmann et al (2010) publication

```python
diekman_R0 = (p*nu_1/(nu_1+mu)+(1-p)*nu_2/(nu_2+mu))*beta/(gamma+mu)
diekman_R0
```

```python
sympy.simplify(sympy_R0 - diekman_R0)==0
```

### 5.6.1 Printing Basic Reproductive Number (R<sub>0</sub>) in latex form for right up.

```python
sympy.print_latex(sympy_R0)
```
